# Pandas Missing Data
- Real world data will often be missing data on variety of reasons
- Many machine learning models and statistical methods can not work with missing data points, in wich case we need to decide what to do with the missing data
- When reading in missing values, pandas will display them as NaN (not a value) values.
- There are also newer specialized null pandas values such as pd.NaT to imply the value missing should be a timestamp

### Options for dealing with missing data:
- Keep it
- Remove it
- Replace it


__Keeping the missing data:__

Pro:
- Easiest to do
- Does not manipulate or change the true data

Cons:
- Many methods do not support NaN
- Often there are reasonable guesses

__Removing (or dropping) the data__

Pro:
- Easy to do
- Can be based on rules

Cons:
- Potential to lose a lot of data or useful information
- Limits trained models for future data

__Replace (filling in) the missing data__

Pros:
- Potential to save a lot of data for use in training a model

Cons:
- Hardest to do and somewhat arbitrary
- Potential to lead to false conclusions
