# .apply() on Single Column
- can use it to apply any costom python function of our onw to every row in a Serie.
- We can use either one or multiple columns as input.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/henkvanhouten/Desktop/Data Science/UNZIP_FOR_NOTEBOOKS_FINAL/03-Pandas/tips.csv")
```

```python
df.head()
```

```python
# last 4 digits creditcard number. In pandas not a method available to convert a int into a str. 
# You have to make one yourself
def last_four(num):
    return str(num) [-4:]
```

```python
df['CC Number'].apply(last_four)
```

```python
# Calculate $, $$, or $$$
# To do so, you have to do some calculations.
# 1. mean
df['total_bill'].mean()
```

```python
# 2. define a function
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
# Create a new column and apply the function
df['yelp'] = df['total_bill'].apply(yelp)
```

```python
df.head()
```

```python
# check rows with '$$'
df[df['yelp'] == '$$']
```

# .apply() on Multiple Columns

# Lambda expression (function without a name)

```python
# normal function:
def simple(num):
    return num * 2
```

```python
# lambda:
lambda num: num * 2
```

```python
df['total_bill'].apply(lambda num: num * 2)  # total_bill * 2
```

```python
df.head()
```

```python
# Tip generous or other
def quality(total_bill, tip):  # same name of column not necessary but good practice
    if tip / total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
# Check the function
quality(16.99, 1.01)
```

```python
# Formatting this function using a lambda expression
df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)  # axis=1 column
```

```python
# assign it to a new column (lambda on df)
df['Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
```

```python
df.head()
```

How to run this a lot faster. And the way to do that is calling NP VECTORISE (faster computational time but also the syntax is a little simpeler)

```python
df["Quality"] = np.vectorize(quality)(df["total_bill"],df['tip'])  # the same, but easier and faster than the lambda
```

```python
df.head()
```
