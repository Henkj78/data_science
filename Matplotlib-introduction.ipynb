## Matplotlib
- Visualizing data is crucial to quickly understanding trends and relationships in your dataset
- Is one of the most popular libraries for plotting with Python.
- Is known as the "Grandfather" of plotting and visualizating libraries for Python
- Many other visualization libraries are built directly of Matplotlib (e.g. seaborn and pandas built-in visualization).
- Is heavily inspired by the plotting functions of the Matlab programming language
- It allows for the creaton of almost any plot type and heavy customization
- There are two approaches to creating plots, functional based methods and OOP bases methods

### Basics
The most basic way to use Matplotlib is throug the function plot calls:
- plt.plot(x, y)

These functions class are simple to use, but don't allow for very high degrees of control
- It is recommended using these simple plt.plot() calls for quickly visualizing relationships and data
- In other cases use the more robust OOP Matplotlib Figure API

### OOP Matplotlib Figure
- The more comprehensive Matplotlib OOP API makes use of a Figure object
- We add axes to this Figure object and then plot on those axes
- This allows for very robust controls over the entire plot

axes = fig.add_axes([0, 0, 1, 1)

Plot inside a plot

```python
fig3 = plt.figure()

axis1 = fig3.add_axes([0,0,1,1])
axis1.plot(a,b)

#Play around with the numbers 
axis2 = fig3.add_axes([0.2,0.2,0.5,0.5])
axis2.plot(x,y);
```

Zoomend in section as an insert

```python
fig3 = plt.figure()

axis = fig3.add_axes([0,0,1,1])
axis.plot(a,b);

# Add axis labels
axis.set_xlim(0,8)
axis.set_ylim(0,8000)
axis.set_xlabel('A')
axis.set_ylabel('B')
axis.set_title("Power of 4")

# set for zoomed in (small plot)
axes1 = fig3.add_axes([0.2,0.5,0.25,0.25])
axes1.plot(a,b)

axes1.set_xlim(1,2)
axes1.set_ylim(0,50)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Zoomed In")
```

```python
fig3 = plt.figure()

axis = fig3.add_axes([0,0,1,1])
axis.plot(a,b);

axis.set_xlim(0,8)
axis.set_ylim(0,8000)
axis.set_xlabel('A')
axis.set_ylabel('B')
axis.set_title("Power of 4")

# zoomed in start from right upper corner
axes1 = fig3.add_axes([1,1,0.25,0.25])
axes1.plot(a,b)

axes1.set_xlim(1,2)
axes1.set_ylim(0,50)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Zoomed In")
```

__Subplots Functionality__
- Matplotlib comes with a pre-configured function called plt.subplots() that automatically create Figure and Axes objects in side by side formations (align multiple plots)
- plt.subplots() returns a tuple which by common convention we label (fig, axes)
- fig: this is the entire Figure canvas
- axes: is a NumPy array holding each of the axes accordin to position in the overall canvas

```python
c = np.linspace(0,10,11)  # evenly linespaced
d = a**4
```

```python
t = np.arange(0,10)
s = 2*t
```

```python
# subplots() default is 1 row by 1 column
fig, axes = plt.subplots()
```

```python
# subplots() add rows and columns (1 row, 2 columns)
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# The moment you start asking for more than one row and column axes is no longer a matplotlib axes object. Instead
# it has become an array which means it has a shape and you can use the array
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(t, s)
axes[1].plot(c, d);
```

```python
# Expand the plots
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0,1].plot(t, s)
axes[1,0].plot(c, d);
```

```python
# under each other (single dimensional array) = used a lot
fig,axes = plt.subplots(nrows=3, ncols=1)

# use the middle one
axes[1].plot(c,d)
```

```python
# Loop through a plot
fig,axes = plt.subplots(nrows=3, ncols=1)

for ax in axes:
    ax.plot(t,s)
```

```python
# adjust spacing automatically
fig,axes = plt.subplots(nrows=3, ncols=1)

for ax in axes:
    ax.plot(t,s)

# Automatically attempt to make sure that set of axes, tickers are overlapping over anything else
plt.tight_layout()
```

```python
# Manually set the spacing
fig,axes = plt.subplots(nrows=2, ncols=2)

axes[0,0].plot(t,s)
axes[0,1].plot(c,d)
axes[1,0].plot(c,d)
axes[1,1].plot(t,s)

# Set spacing manually (1.0 = one axes)
fig.subplots_adjust(wspace=0.3)
fig.subplots_adjust(hspace=0.3);
```

```python
# Exporting the subplots as well as editing the parameters of the subplots
fig,axes = plt.subplots(nrows=2, ncols=2)

axes[0,0].plot(t,s)
axes[0,1].plot(c,d)
axes[1,0].plot(c,d)
axes[1,1].plot(t,s)

# Give title to one of the plots
axes[1,0].set_xlabel("X axes test")
axes[1,0].set_ylabel("Y axes test")
axes[1,0].set_title("Title test")
# google for more

# Title for the entire plot
fig.suptitle("Figure Level", fontsize=16)

# Set spacing manually (1.0 = one axes)
fig.subplots_adjust(wspace=0.3)
fig.subplots_adjust(hspace=0.4);

# Save the entire plot. bbox_inches='tight' = to make sure there is a bounding box that isn't going to cut of anything
fig.savefig("new_subplots.png", bbox_inches='tight')
```

### Matplotlib Styling
- Matplotlib offers very robust styling functions that allow us to edit colors, legends, line width, markers and much more!

__Legends__
- Tells what each line, color etc represents

```python
x = np.linspace(0,11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# line 1 add label
ax.plot(x,x, label="X vs X")
# line 2 add label
ax.plot(x**2, label="X vs X^2")

# Call legend, loc='' default is 'best' or 0. Checks for the best place where is the least amount of information
ax.legend(loc='upper left')

# Manual
ax.legend(loc=(1.1,0.5));  # inside is < 1.0
```

__Visual Styling__

```python
# Color
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# add or change color
ax.plot(x,x,color="lightblue",label='x vs x') 
# custom color
ax.plot(x,x+1,color="#e6c017", label='x vs x + 1');  # RGB HEX Code = formula red/green/blue (google: hex color picker)

ax.legend();
```

```python
# Lines
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# line width 1 = default
ax.plot(x,x,color="lightblue",lw=10) 

# line styles:
# possible linetype options '--', '-', '-.', ':', 'steps'
ax.plot(x,x+1,color="#e6c017",ls='-.');  
```

```python
# Marker
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color="lightblue",lw=5,marker="o", markersize=10,
       markerfacecolor='red', markeredgewidth=8, 
        markeredgecolor='orange');  # tab and scroll down for markers. markersize = ms
```
