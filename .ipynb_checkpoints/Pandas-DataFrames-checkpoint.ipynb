# DataFrames
- A DataFrame is a table of columns and rows in pandas that we can easily restructure and filter
- Formal Definition: A group of Pandas Series objects that share the same index

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(101)
mydata = np.random.randint(0, 101, (4,3))  # up to 101 but not including (4 rows, 3 elements per row)
```

```python
mydata
```

```python
myindex = ['CA', 'NY', 'AZ', 'TX']
```

```python
mycolumns = ['Jan', 'Feb', 'Mar']
```

```python
# Construct a dataframe with only data
df = pd.DataFrame(data=mydata)
```

```python
df
```

```python
# Construct a dataframe with data and index
df = pd.DataFrame(data=mydata, index=myindex)
```

```python
df
```

```python
# Construct a dataframe with data, index and columns
df = pd.DataFrame(data=mydata, index=myindex, columns=mycolumns)
```

```python
df
```

```python
df.info()
```

## Working with columns

```python
df.head()
```

```python
# Grab single column
df['total_bill']
```

```python
# Grab multiple columns
df[['total_bill', 'tip']]
```

```python
# Perform operations 
100 * df['tip'] / df['total_bill']  # percentage of tip
```

```python
df['tip'] + df['total_bill']  # total bill incl tip
```

```python
# Create a new column
df['tip_percentage'] = 100 * df['tip'] / df['total_bill']
```

```python
df.head()
```

```python
# Round tip_percentage to 2 decimal 
df['tip_percentage'] = np.round(100 * df['tip'] / df['total_bill'], 2)
```

```python
df
```

```python
# Removing columns (axis=1 = columns, axis=0 = rows)
df.drop('tip_percentage', axis=1)  # in place (doesn't effect the original array)
```

```python
df  # original array still with tip_percentage
```

```python
# If you want it to be permanent
df = df.drop('tip_percentage', axis=1)
```

```python
df
```

```python
# Insert row (needs to match up with dimensions of current dataframe)
one_row = df.iloc[0]
```

```python
one_row
```

```python
# Concat= concatenating, to_frame()= method to convert one_row into dataframe, .T= to transpose the inde and column
df = pd.concat([df, one_row.to_frame().T])
```
